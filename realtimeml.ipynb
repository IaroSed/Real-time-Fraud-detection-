```
from pyspark.sql.types import *
from pyspark.sql.functions import *


messages = inputStream.withColumn("Offset", col("offset").cast(LongType())) \
                      .withColumn("Timestamp", col("enqueuedTime").cast(TimestampType())) \
                      .withColumn("Body", col("body").cast(StringType())) \
                      .select("Offset", "Timestamp", "Body")

display(messages)
                      
```

```
#Defining schema of the event (transaction)
schema = StructType() \
        .add("Hour", DoubleType()) \
        .add("V1", DoubleType()) \
        .add("V2", DoubleType()) \
        .add("V3", DoubleType()) \
        .add("V4", DoubleType()) \
        .add("V5", DoubleType()) \
        .add("V6", DoubleType()) \
        .add("V7", DoubleType()) \
        .add("V8", DoubleType()) \
        .add("V9", DoubleType()) \
        .add("V10", DoubleType()) \
        .add("V11", DoubleType()) \
        .add("V12", DoubleType()) \
        .add("V13", DoubleType()) \
        .add("V14", DoubleType()) \
        .add("V15", DoubleType()) \
        .add("V16", DoubleType()) \
        .add("V17", DoubleType()) \
        .add("V18", DoubleType()) \
        .add("V19", DoubleType()) \
        .add("V20", DoubleType()) \
        .add("V21", DoubleType()) \
        .add("V22", DoubleType()) \
        .add("V23", DoubleType()) \
        .add("V24", DoubleType()) \
        .add("V25", DoubleType()) \
        .add("V26", DoubleType()) \
        .add("V27", DoubleType()) \
        .add("V28", DoubleType()) \
        .add("Amount", DoubleType())
```

```
#Running a filter query
transactions = messages.select(from_json(col("Body"), schema).alias("data")).select("data.*")
```

```
display(transactions)
```

```
from pyspark.ml.feature import VectorAssembler

#Predicting fraud for the transaction in the stream
columns=['Hour','V1','V2','V3','V4','V5','V6','V7','V8','V9','V10','V11','V12','V13','V14','V15','V16','V17','V18','V19','V20','V21','V22','V23','V24','V25','V26','V27','V28', 'Amount']

assembler = VectorAssembler(inputCols=columns, outputCol='features')
final_transactions= assembler.transform(transactions).select('features')

rfc_preds = rfc_model.transform(final_transactions)
display(rfc_preds.select('prediction'))
```

```
display(rfc_preds.select("features", "prediction").where("prediction = 1"))
```

```
#Saving result stream to disk every 15 seconds
 
#import spark.sql.streaming.Trigger

rfc_preds = rfc_preds \
        .writeStream \
        #.format("delta") \
        .format("csv")
        #.option("checkpintLocation", "/tmp/checkpoint/") \
        .option("path", "/tmp/data/") \
        .trigger(continuous="15 seconds")) \
        .start()
```
